```python
# =====================================================================
# GHOSTNET GROCERY STORE CLASSIFICATION - K-FOLD 5 CROSS VALIDATION
# TANPA AUGMENTASI DATA
# OPTIMIZER: SGD WITH MOMENTUM
# =====================================================================

# =====================================================================
# 1. SETUP AND INSTALLATIONS
# =====================================================================

import os
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
from pathlib import Path
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from tqdm.auto import tqdm
import warnings
import random
warnings.filterwarnings('ignore')

# Install timm for GhostNet
!pip install timm -q

import timm

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    torch.backends.cudnn.deterministic = True

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")

# =====================================================================
# 2. MOUNT GOOGLE DRIVE
# =====================================================================

from google.colab import drive
drive.mount('/content/drive')

# Update this path to your dataset location in Google Drive
DATASET_PATH = '/content/drive/MyDrive/GroceryStoreDataset/GroceryStoreDataset/dataset'

# =====================================================================
# 3. DATASET CLASS - TRAIN
# =====================================================================

class GroceryStoreDataset(Dataset):

    def __init__(self, root_dir, splits=['train'], transform=None):
        self.root_dir = Path(root_dir)
        self.splits = splits if isinstance(splits, list) else [splits]
        self.transform = transform

        # Collect all images and labels
        self.images = []
        self.labels = []
        self.class_names = []

        print(f"\nLoading dataset from: {', '.join(self.splits)}")

        # Collect all leaf folders from all splits
        all_leaf_folders = []
        for split in self.splits:
            data_dir = self.root_dir / split
            if not data_dir.exists():
                print(f"  ⚠️  Warning: {split} folder not found, skipping...")
                continue

            for root, dirs, files in os.walk(data_dir):
                image_files = [f for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
                if image_files and not dirs:
                    all_leaf_folders.append(Path(root))

        # Sort and get unique class names
        all_leaf_folders = sorted(list(set(all_leaf_folders)))

        # Extract class names (leaf folder names)
        class_name_set = set()
        for folder in all_leaf_folders:
            class_name_set.add(folder.name)

        self.class_names = sorted(list(class_name_set))
        self.class_to_idx = {cls_name: idx for idx, cls_name in enumerate(self.class_names)}

        print(f"  ✓ Found {len(self.class_names)} unique classes")

        # Collect all images
        for folder in all_leaf_folders:
            class_name = folder.name
            class_idx = self.class_to_idx[class_name]

            for img_file in folder.glob('*.jpg'):
                self.images.append(str(img_file))
                self.labels.append(class_idx)
            for img_file in folder.glob('*.png'):
                self.images.append(str(img_file))
                self.labels.append(class_idx)

        if len(self.images) > 0:
            unique, counts = np.unique(self.labels, return_counts=True)
            print(f"  ✓ Images per class: min={counts.min()}, max={counts.max()}, avg={counts.mean():.1f}")

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        label = self.labels[idx]

        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            image = Image.new('RGB', (224, 224), (0, 0, 0))

        if self.transform:
            image = self.transform(image)

        return image, label

# =====================================================================
# 4. DATA TRANSFORMATIONS - TANPA AUGMENTASI
# =====================================================================

# Transform yang sama untuk train, validation, dan test (TANPA AUGMENTASI)
standard_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                       std=[0.229, 0.224, 0.225])
])

print("\n⚠️  DATA AUGMENTATION: DISABLED")
print("✓ Using only basic transformations (Resize + CenterCrop + Normalize)")

# =====================================================================
# 5. LOAD DATASETS
# =====================================================================

print("\n" + "="*60)
print("LOADING DATASETS")
print("="*60)

# Load train+val combined
full_trainval_dataset = GroceryStoreDataset(
    root_dir=DATASET_PATH,
    splits=['train'],
    transform=standard_transform
)

# Load test dataset separately
test_dataset = GroceryStoreDataset(
    root_dir=DATASET_PATH,
    splits=['test'],
    transform=standard_transform
)

NUM_CLASSES = len(full_trainval_dataset.class_names)

print(f"\n✓ Total classes: {NUM_CLASSES}")

# Create test loader (same for all folds)
test_loader = DataLoader(
    test_dataset,
    batch_size=64,
    shuffle=False,
    num_workers=0,
    pin_memory=True if torch.cuda.is_available() else False
)

# =====================================================================
# 6. TRANSFORM DATASET CLASS
# =====================================================================

class TransformDataset(Dataset):
    def __init__(self, subset, transform):
        self.subset = subset
        self.transform = transform

    def __len__(self):
        return len(self.subset)

    def __getitem__(self, idx):
        img_path = self.subset.dataset.images[self.subset.indices[idx]]
        label = self.subset.dataset.labels[self.subset.indices[idx]]

        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            image = Image.new('RGB', (224, 224), (0, 0, 0))

        if self.transform:
            image = self.transform(image)

        return image, label

# =====================================================================
# 7. TRAINING AND EVALUATION FUNCTIONS
# =====================================================================

def train_epoch(model, train_loader, criterion, optimizer, device):
    """Train for one epoch"""
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    pbar = tqdm(train_loader, desc='Training', leave=False)
    for inputs, labels in pbar:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        pbar.set_postfix({
            'loss': f'{loss.item():.4f}',
            'acc': f'{100.*correct/total:.2f}%'
        })

    epoch_loss = running_loss / len(train_loader.dataset)
    epoch_acc = 100. * correct / total

    return epoch_loss, epoch_acc

def evaluate(model, loader, criterion, device):
    """Evaluate the model"""
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        pbar = tqdm(loader, desc='Evaluating', leave=False)
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            pbar.set_postfix({
                'loss': f'{loss.item():.4f}',
                'acc': f'{100.*correct/total:.2f}%'
            })

    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = 100. * correct / total

    return epoch_loss, epoch_acc, all_preds, all_labels

def evaluate_detailed(model, loader, device):
    """Evaluate model with detailed metrics"""
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        pbar = tqdm(loader, desc='Test Evaluation', leave=False)
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds) * 100
    precision = precision_score(all_labels, all_preds, average='weighted', zero_division=0) * 100
    recall = recall_score(all_labels, all_preds, average='weighted', zero_division=0) * 100
    f1 = f1_score(all_labels, all_preds, average='weighted', zero_division=0) * 100

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'predictions': all_preds,
        'labels': all_labels
    }

# =====================================================================
# 8. K-FOLD CROSS VALIDATION SETUP
# =====================================================================

BATCH_SIZE = 64
NUM_WORKERS = 0
EPOCHS = 10
INITIAL_LR = 0.01
MOMENTUM = 0.9
WEIGHT_DECAY = 1e-4
N_FOLDS = 5

print("\n" + "="*60)
print("K-FOLD CROSS VALIDATION CONFIGURATION")
print("="*60)
print(f"✓ Number of folds: {N_FOLDS}")
print(f"✓ Epochs per fold: {EPOCHS}")
print(f"✓ Batch size: {BATCH_SIZE}")
print(f"✓ Initial LR: {INITIAL_LR}")
print(f"✓ Momentum: {MOMENTUM}")
print(f"✓ Weight Decay: {WEIGHT_DECAY}")
print(f"✓ Optimizer: SGD")
print(f"✓ Model: GhostNet-100")
print(f"⚠️  Data Augmentation: DISABLED")

# Setup K-Fold
indices = list(range(len(full_trainval_dataset)))
labels = full_trainval_dataset.labels

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=42)

# Store results for all folds
fold_histories = []
fold_test_evaluations = []

# =====================================================================
# 9. K-FOLD TRAINING LOOP WITH TEST SET EVALUATION
# =====================================================================

print("\n" + "="*60)
print("STARTING K-FOLD CROSS VALIDATION (NO AUGMENTATION)")
print("="*60)

for fold, (train_idx, val_idx) in enumerate(skf.split(indices, labels)):
    print(f"\n{'='*60}")
    print(f"FOLD {fold+1}/{N_FOLDS}")
    print(f"{'='*60}")

    # Create train and val subsets for this fold
    train_subset = Subset(full_trainval_dataset, train_idx)
    val_subset = Subset(full_trainval_dataset, val_idx)

    # Apply same transform to both (NO AUGMENTATION)
    train_dataset_fold = TransformDataset(train_subset, standard_transform)
    val_dataset_fold = TransformDataset(val_subset, standard_transform)

    # Create dataloaders
    train_loader = DataLoader(
        train_dataset_fold,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=NUM_WORKERS,
        pin_memory=True if torch.cuda.is_available() else False
    )

    val_loader = DataLoader(
        val_dataset_fold,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=NUM_WORKERS,
        pin_memory=True if torch.cuda.is_available() else False
    )

    # Initialize model
    model = timm.create_model('ghostnet_100', pretrained=True)
    num_ftrs = model.classifier.in_features
    print(f"✓ Feature dimension: {num_ftrs}")

    model.classifier = nn.Linear(num_ftrs, NUM_CLASSES)
    nn.init.normal_(model.classifier.weight, mean=0.0, std=0.01)
    nn.init.constant_(model.classifier.bias, 0)
    model = model.to(device)

    # Loss and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(
        model.parameters(),
        lr=INITIAL_LR,
        momentum=MOMENTUM,
        weight_decay=WEIGHT_DECAY,
        nesterov=True
    )

    scheduler = optim.lr_scheduler.StepLR(
        optimizer,
        step_size=10,
        gamma=0.1
    )

    # Training history for this fold
    fold_history = {
        'train_loss': [],
        'train_acc': [],
        'val_loss': [],
        'val_acc': [],
        'lr': []
    }

    best_val_acc = 0.0
    best_epoch = 0

    # Train for this fold
    for epoch in range(EPOCHS):
        current_lr = optimizer.param_groups[0]['lr']

        print(f"\nFold {fold+1} - Epoch {epoch+1}/{EPOCHS} | LR: {current_lr:.6f}")
        print("-" * 60)

        # Train
        train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)

        # Validate
        val_loss, val_acc, _, _ = evaluate(model, val_loader, criterion, device)

        # Step scheduler
        old_lr = current_lr
        scheduler.step()
        new_lr = optimizer.param_groups[0]['lr']

        if new_lr < old_lr:
            print(f"📉 Learning rate reduced: {old_lr:.6f} → {new_lr:.6f}")

        # Save history
        fold_history['train_loss'].append(train_loss)
        fold_history['train_acc'].append(train_acc)
        fold_history['val_loss'].append(val_loss)
        fold_history['val_acc'].append(val_acc)
        fold_history['lr'].append(current_lr)

        # Print results
        print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
        print(f"Val Loss:   {val_loss:.4f} | Val Acc:   {val_acc:.2f}%")

        # Track best validation accuracy
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_epoch = epoch + 1
            print(f"✓ New best validation accuracy! (Val Acc: {best_val_acc:.2f}%)")

    # =====================================================================
    # EVALUATE THE FINAL MODEL ON TEST SET
    # =====================================================================
    print(f"\n{'='*60}")
    print(f"EVALUATING FOLD {fold+1} ON TEST SET")
    print(f"{'='*60}")

    test_results = evaluate_detailed(model, test_loader, device)

    print(f"\n📊 Test Set Evaluation Results (Fold {fold+1}):")
    print(f"  • Accuracy:  {test_results['accuracy']:.2f}%")
    print(f"  • Precision: {test_results['precision']:.2f}%")
    print(f"  • Recall:    {test_results['recall']:.2f}%")
    print(f"  • F1-Score:  {test_results['f1_score']:.2f}%")

    # Save model with TEST results
    fold_model_path = f'/content/drive/MyDrive/510fold{fold+1}_final_model_ghostnet_sgd_noaug.pth'
    torch.save({
        'fold': fold + 1,
        'epoch': EPOCHS,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'num_classes': NUM_CLASSES,
        'num_ftrs': num_ftrs,
        'test_evaluation_metrics': {
            'accuracy': test_results['accuracy'],
            'precision': test_results['precision'],
            'recall': test_results['recall'],
            'f1_score': test_results['f1_score']
        },
        'best_val_acc': best_val_acc,
        'best_val_epoch': best_epoch,
        'augmentation': False
    }, fold_model_path)
    print(f"✓ Model saved: {fold_model_path}")

    # Store fold TEST evaluation results
    fold_test_evaluations.append({
        'fold': fold + 1,
        'test_accuracy': test_results['accuracy'],
        'test_precision': test_results['precision'],
        'test_recall': test_results['recall'],
        'test_f1_score': test_results['f1_score'],
        'best_val_acc': best_val_acc,
        'best_val_epoch': best_epoch,
        'model_path': fold_model_path
    })

    # Store fold training history
    fold_histories.append(fold_history)

    print(f"\nFold {fold+1} Summary:")
    print(f"  • Val Acc: {best_val_acc:.2f}%")
    print(f"  • Test Accuracy: {test_results['accuracy']:.2f}%")
    print(f"  • Test F1-Score: {test_results['f1_score']:.2f}%")

# =====================================================================
# 10. K-FOLD TEST RESULTS SUMMARY
# =====================================================================

print("\n" + "="*60)
print("K-FOLD TEST SET EVALUATION RESULTS (NO AUGMENTATION)")
print("="*60)

# Create DataFrame for better visualization
test_eval_df = pd.DataFrame(fold_test_evaluations)

print("\nDetailed Test Results Per Fold:")
print(test_eval_df[['fold', 'test_accuracy', 'test_precision', 'test_recall', 'test_f1_score']].to_string(index=False))

print("\n" + "="*60)
print("SUMMARY STATISTICS (TEST SET)")
print("="*60)

# Calculate mean and std for each metric
test_metrics = ['test_accuracy', 'test_precision', 'test_recall', 'test_f1_score']
summary_stats = {}

for metric in test_metrics:
    values = test_eval_df[metric].values
    mean_val = np.mean(values)
    std_val = np.std(values)
    summary_stats[metric] = {'mean': mean_val, 'std': std_val}

    metric_name = metric.replace('test_', '').replace('_', ' ').title()
    print(f"\n{metric_name}:")
    print(f"  • Mean: {mean_val:.2f}%")
    print(f"  • Std:  {std_val:.2f}%")
    print(f"  • Min:  {np.min(values):.2f}%")
    print(f"  • Max:  {np.max(values):.2f}%")

# =====================================================================
# 11. SAVE RESULTS TO CSV
# =====================================================================

test_eval_df.to_csv('kfold_test_evaluation_results_noaug.csv', index=False)
print("\n✓ Test results saved to 'kfold_test_evaluation_results_noaug.csv'")

# Save summary statistics
summary_df = pd.DataFrame(summary_stats).T
summary_df.to_csv('kfold_test_summary_statistics_noaug.csv')
print("✓ Test summary statistics saved to 'kfold_test_summary_statistics_noaug.csv'")

# =====================================================================
# 12. PLOT K-FOLD TEST EVALUATION RESULTS
# =====================================================================

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Test Accuracy per fold
ax = axes[0, 0]
folds = test_eval_df['fold'].values
accuracies = test_eval_df['test_accuracy'].values
ax.bar(folds, accuracies, color='steelblue', alpha=0.7)
ax.axhline(summary_stats['test_accuracy']['mean'], color='red', linestyle='--',
           label=f"Mean: {summary_stats['test_accuracy']['mean']:.2f}%")
ax.fill_between([0.5, N_FOLDS+0.5],
                summary_stats['test_accuracy']['mean'] - summary_stats['test_accuracy']['std'],
                summary_stats['test_accuracy']['mean'] + summary_stats['test_accuracy']['std'],
                alpha=0.2, color='red', label=f"±1 Std: {summary_stats['test_accuracy']['std']:.2f}%")
ax.set_xlabel('Fold')
ax.set_ylabel('Test Accuracy (%)')
ax.set_title('Test Accuracy per Fold (No Augmentation)')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
ax.set_xticks(folds)

# Plot 2: Test Precision per fold
ax = axes[0, 1]
precisions = test_eval_df['test_precision'].values
ax.bar(folds, precisions, color='green', alpha=0.7)
ax.axhline(summary_stats['test_precision']['mean'], color='red', linestyle='--',
           label=f"Mean: {summary_stats['test_precision']['mean']:.2f}%")
ax.fill_between([0.5, N_FOLDS+0.5],
                summary_stats['test_precision']['mean'] - summary_stats['test_precision']['std'],
                summary_stats['test_precision']['mean'] + summary_stats['test_precision']['std'],
                alpha=0.2, color='red', label=f"±1 Std: {summary_stats['test_precision']['std']:.2f}%")
ax.set_xlabel('Fold')
ax.set_ylabel('Test Precision (%)')
ax.set_title('Test Precision per Fold (No Augmentation)')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
ax.set_xticks(folds)

# Plot 3: Test Recall per fold
ax = axes[1, 0]
recalls = test_eval_df['test_recall'].values
ax.bar(folds, recalls, color='orange', alpha=0.7)
ax.axhline(summary_stats['test_recall']['mean'], color='red', linestyle='--',
           label=f"Mean: {summary_stats['test_recall']['mean']:.2f}%")
ax.fill_between([0.5, N_FOLDS+0.5],
                summary_stats['test_recall']['mean'] - summary_stats['test_recall']['std'],
                summary_stats['test_recall']['mean'] + summary_stats['test_recall']['std'],
                alpha=0.2, color='red', label=f"±1 Std: {summary_stats['test_recall']['std']:.2f}%")
ax.set_xlabel('Fold')
ax.set_ylabel('Test Recall (%)')
ax.set_title('Test Recall per Fold (No Augmentation)')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
ax.set_xticks(folds)

# Plot 4: Test F1-Score per fold
ax = axes[1, 1]
f1_scores = test_eval_df['test_f1_score'].values
ax.bar(folds, f1_scores, color='purple', alpha=0.7)
ax.axhline(summary_stats['test_f1_score']['mean'], color='red', linestyle='--',
           label=f"Mean: {summary_stats['test_f1_score']['mean']:.2f}%")
ax.fill_between([0.5, N_FOLDS+0.5],
                summary_stats['test_f1_score']['mean'] - summary_stats['test_f1_score']['std'],
                summary_stats['test_f1_score']['mean'] + summary_stats['test_f1_score']['std'],
                alpha=0.2, color='red', label=f"±1 Std: {summary_stats['test_f1_score']['std']:.2f}%")
ax.set_xlabel('Fold')
ax.set_ylabel('Test F1-Score (%)')
ax.set_title('Test F1-Score per Fold (No Augmentation)')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
ax.set_xticks(folds)

plt.tight_layout()
plt.savefig('kfold_test_evaluation_metrics_noaug.png', dpi=300, bbox_inches='tight')
plt.show()

# =====================================================================
# 13. PLOT ALL TEST METRICS COMPARISON
# =====================================================================

fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(N_FOLDS)
width = 0.2

ax.bar(x - 1.5*width, test_eval_df['test_accuracy'], width, label='Accuracy', color='steelblue', alpha=0.8)
ax.bar(x - 0.5*width, test_eval_df['test_precision'], width, label='Precision', color='green', alpha=0.8)
ax.bar(x + 0.5*width, test_eval_df['test_recall'], width, label='Recall', color='orange', alpha=0.8)
ax.bar(x + 1.5*width, test_eval_df['test_f1_score'], width, label='F1-Score', color='purple', alpha=0.8)

ax.set_xlabel('Fold', fontsize=12)
ax.set_ylabel('Test Score (%)', fontsize=12)
ax.set_title('All Test Metrics Comparison Across Folds (No Augmentation)', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels([f'Fold {i+1}' for i in range(N_FOLDS)])
ax.legend()
ax.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('kfold_test_all_metrics_comparison_noaug.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("K-FOLD CROSS VALIDATION COMPLETE! (NO AUGMENTATION)")
print("="*60)
print("✓ All plots saved")
print("✓ Results saved to CSV files")
print(f"\n✓ Saved {N_FOLDS} model files with TEST evaluations:")
for i in range(1, N_FOLDS+1):
    print(f"  - fold{i}_final_model_ghostnet_sgd_noaug.pth")

print(f"\n📊 Final Test Set Performance (NO AUGMENTATION):")
print(f"  • Mean Accuracy:  {summary_stats['test_accuracy']['mean']:.2f}% ± {summary_stats['test_accuracy']['std']:.2f}%")
print(f"  • Mean Precision: {summary_stats['test_precision']['mean']:.2f}% ± {summary_stats['test_precision']['std']:.2f}%")
print(f"  • Mean Recall:    {summary_stats['test_recall']['mean']:.2f}% ± {summary_stats['test_recall']['std']:.2f}%")
print(f"  • Mean F1-Score:  {summary_stats['test_f1_score']['mean']:.2f}% ± {summary_stats['test_f1_score']['std']:.2f}%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.2)
sns.set_palette("husl")

# ============================================================
# EXTRACTED DATA FROM TRAINING LOGS (NO AUGMENTATION)
# ============================================================

training_data = {
    'fold1': {
        'epochs': list(range(1, 11)),
        'train_loss': [4.0755, 2.3737, 0.7519, 0.2822, 0.1445, 0.0774, 0.0505, 0.0449, 0.0295, 0.0233],
        'val_loss': [3.4152, 1.3690, 0.4989, 0.2947, 0.2284, 0.1888, 0.1715, 0.1542, 0.1377, 0.1316],
        'train_acc': [18.44, 55.83, 88.80, 95.19, 98.51, 99.62, 99.87, 99.74, 99.91, 99.96],
        'val_acc': [38.95, 79.08, 90.14, 93.03, 94.56, 95.92, 95.07, 95.75, 96.43, 96.26],
        'test_acc': 85.31
    },
    'fold2': {
        'epochs': list(range(1, 11)),
        'train_loss': [4.0809, 2.4253, 0.8245, 0.2889, 0.1432, 0.0853, 0.0541, 0.0394, 0.0346, 0.0266],
        'val_loss': [3.3996, 1.3418, 0.4453, 0.2666, 0.1644, 0.1354, 0.1154, 0.1027, 0.0932, 0.0858],
        'train_acc': [17.79, 55.64, 85.65, 95.19, 98.25, 99.45, 99.74, 99.79, 99.83, 99.91],
        'val_acc': [41.06, 75.47, 91.14, 93.87, 96.25, 96.93, 97.61, 97.96, 97.61, 97.79],
        'test_acc': 85.15
    },
    'fold3': {
        'epochs': list(range(1, 11)),
        'train_loss': [4.0827, 2.4099, 0.7935, 0.2838, 0.1470, 0.0900, 0.0543, 0.0426, 0.0294, 0.0266],
        'val_loss': [3.3876, 1.3503, 0.4528, 0.2763, 0.1905, 0.1627, 0.1405, 0.1336, 0.1239, 0.1204],
        'train_acc': [16.65, 56.02, 86.97, 95.66, 98.13, 99.36, 99.83, 99.83, 100.00, 99.87],
        'val_acc': [35.26, 74.62, 93.02, 95.40, 96.76, 97.44, 97.79, 96.76, 97.10, 96.93],
        'test_acc': 84.51
    },
    'fold4': {
        'epochs': list(range(1, 11)),
        'train_loss': [4.0891, 2.4055, 0.7726, 0.2605, 0.1307, 0.0772, 0.0562, 0.0389, 0.0311, 0.0301],
        'val_loss': [3.4306, 1.4285, 0.5347, 0.3377, 0.2817, 0.2498, 0.2072, 0.1968, 0.1914, 0.1910],
        'train_acc': [17.11, 56.19, 88.29, 96.21, 98.68, 99.83, 99.66, 99.96, 99.91, 99.74],
        'val_acc': [34.41, 73.42, 88.59, 91.65, 92.84, 93.53, 95.57, 95.06, 94.72, 95.23],
        'test_acc': 85.96
    },
    'fold5': {
        'epochs': list(range(1, 11)),
        'train_loss': [4.0756, 2.4061, 0.8013, 0.2895, 0.1534, 0.0839, 0.0549, 0.0380, 0.0302, 0.0255],
        'val_loss': [3.4343, 1.3613, 0.4987, 0.2864, 0.2016, 0.1641, 0.1497, 0.1356, 0.1278, 0.1222],
        'train_acc': [17.37, 57.30, 87.36, 95.27, 97.74, 99.15, 99.70, 99.91, 99.96, 99.96],
        'val_acc': [39.18, 77.68, 89.61, 93.53, 96.59, 96.93, 96.59, 96.59, 97.10, 96.59],
        'test_acc': 84.23
    }
}

# ============================================================
# 1. VALIDATION ERROR (VAL LOSS) - ALL FOLDS
# ============================================================

fig, ax = plt.subplots(figsize=(14, 8))

# Plot individual folds
colors = sns.color_palette("husl", 5)
for idx, (fold_name, fold_data) in enumerate(training_data.items()):
    ax.plot(fold_data['epochs'], fold_data['val_loss'],
            marker='o', linewidth=2, alpha=0.7,
            label=f'{fold_name.upper()}', color=colors[idx])

# Calculate and plot average
epochs = list(range(1, 11))
avg_val_loss = [np.mean([training_data[f'fold{i}']['val_loss'][e-1] for i in range(1, 6)])
                for e in epochs]
ax.plot(epochs, avg_val_loss,
        marker='s', linewidth=3.5, color='black',
        label='AVERAGE', linestyle='--', markersize=8)

ax.set_xlabel('Epoch', fontsize=14, fontweight='bold')
ax.set_ylabel('Validation Loss', fontsize=14, fontweight='bold')
ax.set_title('📉 Validation Error (Loss) - All Folds (NO AUGMENTATION)\nGhostNet K-Fold Cross Validation',
             fontsize=16, fontweight='bold', pad=20)
ax.legend(loc='upper right', frameon=True, shadow=True, fontsize=11)
ax.grid(True, alpha=0.3)

# Add final value annotation
final_val_loss = avg_val_loss[-1]
ax.annotate(f'Final Avg: {final_val_loss:.4f}',
            xy=(10, final_val_loss), xytext=(8.5, final_val_loss + 0.3),
            arrowprops=dict(arrowstyle='->', color='red', lw=2),
            fontsize=12, fontweight='bold', color='red',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7))

plt.tight_layout()
plt.savefig('validation_error_curves_noaug.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"✅ Final Average Validation Loss: {final_val_loss:.4f}")
print(f"📉 Loss Reduction: {avg_val_loss[0]:.4f} → {final_val_loss:.4f} ({(1 - final_val_loss/avg_val_loss[0])*100:.2f}% reduction)")

# ============================================================
# 2. TRAIN VS TEST ACCURACY - PER FOLD
# ============================================================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

# Prepare data
folds = [f'Fold {i}' for i in range(1, 6)]
train_acc_final = [training_data[f'fold{i}']['train_acc'][-1] for i in range(1, 6)]
test_acc = [training_data[f'fold{i}']['test_acc'] for i in range(1, 6)]
gaps = [train_acc_final[i] - test_acc[i] for i in range(5)]

# Bar chart comparison
x = np.arange(len(folds))
width = 0.35

bars1 = ax1.bar(x - width/2, train_acc_final, width, label='Train Acc (Epoch 10)',
                color=sns.color_palette("Greens", 3)[2], edgecolor='black', linewidth=1.5)
bars2 = ax1.bar(x + width/2, test_acc, width, label='Test Acc',
                color=sns.color_palette("Blues", 3)[2], edgecolor='black', linewidth=1.5)

ax1.set_xlabel('Fold', fontsize=14, fontweight='bold')
ax1.set_ylabel('Accuracy (%)', fontsize=14, fontweight='bold')
ax1.set_title('⚖️ Train vs Test Accuracy - Per Fold (NO AUGMENTATION)', fontsize=16, fontweight='bold', pad=15)
ax1.set_xticks(x)
ax1.set_xticklabels(folds)
ax1.legend(frameon=True, shadow=True, fontsize=12)
ax1.set_ylim([80, 102])
ax1.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.2f}%', ha='center', va='bottom', fontsize=9, fontweight='bold')

# Overfitting gap scatter plot
ax2.scatter(range(1, 6), gaps, s=200, c=gaps, cmap='YlOrRd',
           edgecolors='black', linewidth=2, alpha=0.8)
ax2.plot(range(1, 6), gaps, linestyle='--', color='gray', alpha=0.5, linewidth=2)

# Add horizontal line for average gap
avg_gap = np.mean(gaps)
ax2.axhline(y=avg_gap, color='red', linestyle='--', linewidth=2.5,
           label=f'Avg Gap: {avg_gap:.2f}%')

ax2.set_xlabel('Fold Number', fontsize=14, fontweight='bold')
ax2.set_ylabel('Train-Test Gap (%)', fontsize=14, fontweight='bold')
ax2.set_title('🎯 Overfitting Gap Analysis (NO AUGMENTATION)', fontsize=16, fontweight='bold', pad=15)
ax2.set_xticks(range(1, 6))
ax2.legend(frameon=True, shadow=True, fontsize=12)
ax2.grid(True, alpha=0.3)

# Add colorbar
cbar = plt.colorbar(ax2.collections[0], ax=ax2)
cbar.set_label('Gap Magnitude', rotation=270, labelpad=20, fontweight='bold')

plt.tight_layout()
plt.savefig('train_vs_test_accuracy_noaug.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n✅ Average Train Accuracy (Final): {np.mean(train_acc_final):.2f}%")
print(f"✅ Average Test Accuracy: {np.mean(test_acc):.2f}%")
print(f"⚠️  Average Overfitting Gap: {avg_gap:.2f}%")

# ============================================================
# 3. AVERAGE TRAIN VS VAL LOSS
# ============================================================

fig, ax = plt.subplots(figsize=(14, 8))

# Calculate averages
avg_train_loss = [np.mean([training_data[f'fold{i}']['train_loss'][e-1] for i in range(1, 6)])
                  for e in epochs]
avg_val_loss = [np.mean([training_data[f'fold{i}']['val_loss'][e-1] for i in range(1, 6)])
                for e in epochs]

# Plot
ax.plot(epochs, avg_train_loss, marker='o', linewidth=3,
        label='Train Loss', color=sns.color_palette("Greens", 3)[2], markersize=10)
ax.plot(epochs, avg_val_loss, marker='s', linewidth=3,
        label='Validation Loss', color=sns.color_palette("Reds", 3)[2], markersize=10)

# Fill between
ax.fill_between(epochs, avg_train_loss, avg_val_loss,
                alpha=0.2, color='orange', label='Gap')

ax.set_xlabel('Epoch', fontsize=14, fontweight='bold')
ax.set_ylabel('Loss', fontsize=14, fontweight='bold')
ax.set_title('📊 Average Train vs Validation Loss (NO AUGMENTATION)\nGhostNet-100 K-Fold Cross Validation',
             fontsize=16, fontweight='bold', pad=20)
ax.legend(loc='upper right', frameon=True, shadow=True, fontsize=12)
ax.grid(True, alpha=0.3)

# Add annotations
ax.annotate(f'Start: {avg_train_loss[0]:.4f}',
            xy=(1, avg_train_loss[0]), xytext=(2, avg_train_loss[0] + 0.5),
            arrowprops=dict(arrowstyle='->', color='green', lw=2),
            fontsize=11, fontweight='bold', color='green')

ax.annotate(f'End: {avg_train_loss[-1]:.4f}',
            xy=(10, avg_train_loss[-1]), xytext=(8, avg_train_loss[-1] + 0.3),
            arrowprops=dict(arrowstyle='->', color='green', lw=2),
            fontsize=11, fontweight='bold', color='green')

plt.tight_layout()
plt.savefig('loss_comparison_noaug.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n✅ Train Loss: {avg_train_loss[0]:.4f} → {avg_train_loss[-1]:.4f} ({(1 - avg_train_loss[-1]/avg_train_loss[0])*100:.2f}% reduction)")
print(f"✅ Val Loss: {avg_val_loss[0]:.4f} → {avg_val_loss[-1]:.4f} ({(1 - avg_val_loss[-1]/avg_val_loss[0])*100:.2f}% reduction)")

# ============================================================
# 4. AVERAGE TRAIN VS VAL ACCURACY
# ============================================================

fig, ax = plt.subplots(figsize=(14, 8))

# Calculate averages
avg_train_acc = [np.mean([training_data[f'fold{i}']['train_acc'][e-1] for i in range(1, 6)])
                 for e in epochs]
avg_val_acc = [np.mean([training_data[f'fold{i}']['val_acc'][e-1] for i in range(1, 6)])
               for e in epochs]

# Plot
ax.plot(epochs, avg_train_acc, marker='o', linewidth=3,
        label='Train Accuracy', color=sns.color_palette("Greens", 3)[2], markersize=10)
ax.plot(epochs, avg_val_acc, marker='s', linewidth=3,
        label='Validation Accuracy', color=sns.color_palette("Blues", 3)[2], markersize=10)

# Fill between
ax.fill_between(epochs, avg_train_acc, avg_val_acc,
                alpha=0.2, color='orange', label='Gap')

ax.set_xlabel('Epoch', fontsize=14, fontweight='bold')
ax.set_ylabel('Accuracy (%)', fontsize=14, fontweight='bold')
ax.set_title('📈 Average Train vs Validation Accuracy (NO AUGMENTATION)\nGhostNet K-Fold Cross Validation',
             fontsize=16, fontweight='bold', pad=20)
ax.legend(loc='lower right', frameon=True, shadow=True, fontsize=12)
ax.set_ylim([0, 105])
ax.grid(True, alpha=0.3)

# Add annotations
ax.annotate(f'Final Train: {avg_train_acc[-1]:.2f}%',
            xy=(10, avg_train_acc[-1]), xytext=(7, avg_train_acc[-1] - 8),
            arrowprops=dict(arrowstyle='->', color='green', lw=2),
            fontsize=11, fontweight='bold', color='green',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgreen', alpha=0.7))

ax.annotate(f'Final Val: {avg_val_acc[-1]:.2f}%',
            xy=(10, avg_val_acc[-1]), xytext=(7, avg_val_acc[-1] + 3),
            arrowprops=dict(arrowstyle='->', color='blue', lw=2),
            fontsize=11, fontweight='bold', color='blue',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='lightblue', alpha=0.7))

plt.tight_layout()
plt.savefig('accuracy_comparison_noaug.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n✅ Final Train Accuracy: {avg_train_acc[-1]:.2f}%")
print(f"✅ Final Validation Accuracy: {avg_val_acc[-1]:.2f}%")
print(f"⚠️  Train-Val Gap: {avg_train_acc[-1] - avg_val_acc[-1]:.2f}%")

# ============================================================
# 5. SUMMARY HEATMAP
# ============================================================

fig, ax = plt.subplots(figsize=(12, 8))

# Prepare summary data
summary_data = []
for i in range(1, 6):
    fold_data = training_data[f'fold{i}']
    summary_data.append([
        fold_data['train_acc'][-1],  # Final train acc
        fold_data['val_acc'][-1],    # Final val acc
        fold_data['test_acc'],       # Test acc
        fold_data['train_loss'][-1], # Final train loss
        fold_data['val_loss'][-1],   # Final val loss
    ])

summary_df = pd.DataFrame(summary_data,
                          columns=['Train Acc', 'Val Acc', 'Test Acc', 'Train Loss', 'Val Loss'],
                          index=[f'Fold {i}' for i in range(1, 6)])

# Add average row
avg_row = summary_df.mean()
summary_df.loc['AVERAGE'] = avg_row

# Create heatmap
sns.heatmap(summary_df, annot=True, fmt='.2f', cmap='RdYlGn',
            linewidths=2, linecolor='black', cbar_kws={'label': 'Value'},
            vmin=0, vmax=100, ax=ax)

ax.set_title('📊 Summary Heatmap - All Metrics by Fold (NO AUGMENTATION)\nGhostNet-100 K-Fold Results',
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Metrics', fontsize=13, fontweight='bold')
ax.set_ylabel('Fold', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('summary_heatmap_noaug.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("📊 SUMMARY STATISTICS (NO AUGMENTATION)")
print("="*60)
print(summary_df.round(2))
print("\n✅ All visualizations saved successfully!")
print("   • validation_error_curves_noaug.png")
print("   • train_vs_test_accuracy_noaug.png")
print("   • loss_comparison_noaug.png")
print("   • accuracy_comparison_noaug.png")
print("   • summary_heatmap_noaug.png")
```
